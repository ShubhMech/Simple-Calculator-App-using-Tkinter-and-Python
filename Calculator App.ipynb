```python
from tkinter import *
```

```python
root=Tk()
root.title("Calculator")

entry=Entry(root, width=35, borderwidth=5)
# entry.insert(0,"Make your entry")
entry.grid(row=0,column=0, columnspan=3, padx=10, pady=10)

def button_clear():
    entry.delete(0,END)

def button_add(number):
    current=entry.get()
    
    entry.delete(0,END)
    entry.insert(0,current+str(number))

    
def button_addition(number):
    current=entry.get()
    entry.delete(0,END)
    entry.insert(0,current+str(number))
    global prev
    prev=float(current)
    global sign
    sign="+"
    
def button_subtraction(number):
    current=entry.get()
    entry.delete(0,END)
    entry.insert(0,current+str(number))
    global prev
    prev=float(current)
    global sign
    sign="-"
    
def button_division(number):
    current=entry.get()
    entry.delete(0,END)
    entry.insert(0,current+str(number))
    global prev
    prev=float(current)
    global sign
    sign="/"
    
def button_multiplication(number):
    current=entry.get()
    entry.delete(0,END)
    entry.insert(0,current+str(number))
    global prev
    prev=float(current)
    global sign
    sign="*"
    


def equals_to():
    global sign
    current=entry.get()
#     indice=current.index(sign)
    
#     aa[aa.index("+")+1:]
#     current=current[indice+1:]
    if sign=="+":
        indice=current.index("+")
        current=current[indice+1:]
        result=prev+float(current)
        entry.delete(0,END)
    elif sign=="-":
        indice=current.index("-")
        current=current[indice+1:]
        result=prev-float(current)
        entry.delete(0,END)
    elif sign=="/":
        indice=current.index("/")
        current=current[indice+1:]
        if float(current) != 0:
            result=prev/float(current)
            entry.delete(0,END)
        else:
            entry.insert(0,"NA")
    elif sign=="*":
        indice=current.index("*")
        current=current[indice+1:]
        result=prev*float(current)
        entry.delete(0,END)
        
    entry.insert(0,result)
    
    

button1=Button(root, text="1", command=lambda:button_add(1), padx=40, pady=20)
button1.grid(row=3, column=0)
button2=Button(root, text="2", command=lambda:button_add(2), padx=40, pady=20)
button2.grid(row=3, column=1)
button3=Button(root, text="3", command=lambda:button_add(3), padx=40, pady=20)
button3.grid(row=3, column=2)
button4=Button(root, text="4", command=lambda:button_add(4), padx=40, pady=20)
button4.grid(row=2, column=0)
button5=Button(root, text="5", command=lambda:button_add(5), padx=40, pady=20)
button5.grid(row=2, column=1)
button6=Button(root, text="6", command=lambda:button_add(6), padx=40, pady=20)
button6.grid(row=2, column=2)
button7=Button(root, text="7", command=lambda:button_add(7), padx=40, pady=20)
button7.grid(row=1, column=0)
button8=Button(root, text="8", command=lambda:button_add(8), padx=40, pady=20)
button8.grid(row=1, column=1)
button9=Button(root, text="9", command=lambda:button_add(9), padx=40, pady=20)
button9.grid(row=1, column=2)
button0=Button(root, text="0", command=lambda:button_add(0), padx=40, pady=20)
button0.grid(row=4, column=0)
button_dot=Button(root, text=".", command=lambda:button_add("."), padx=41.5, pady=20)
button_dot.grid(row=4, column=1)
button_plus=Button(root, text="+", command=lambda: button_addition("+"), padx=40, pady=20)
button_plus.grid(row=4, column=2)

button_plus=Button(root, text="-", command=lambda: button_subtraction("-"), padx=40, pady=20)
button_plus.grid(row=5, column=0)
button_plus=Button(root, text="/", command=lambda: button_division("/"), padx=40, pady=20)
button_plus.grid(row=5, column=1)
button_plus=Button(root, text="*", command=lambda: button_multiplication("*"), padx=40, pady=20)
button_plus.grid(row=5, column=2)

button_equals=Button(root, text="=", command=equals_to, padx=88, pady=20)
button_equals.grid(row=6, column=0, columnspan=2)

button_clear=Button(root, text="Clear", command=button_clear, padx=40, pady=20)
button_clear.grid(row=6, column=2)

# button7=Button(root, text="7", command=button_add, padx=40, pady=20)
# button7.grid(row=1, column=0)

root.mainloop()
```
